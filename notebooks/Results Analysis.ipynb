```python
import pandas as pd
import numpy as np
```

```python
def build_table(modes = np.arange(1402,2830,1)):
    df_final = pd.DataFrame()
    fails = []
    for mode in modes:        
        try: 
            filename = 'results_2_23/results_M3F'+str(mode)+'.csv'
            df_new = pd.read_csv(filename)
            #df = pd.concat([df, df_new], axis = 1)
            df_final = pd.concat([df_final, df_new], axis = 0, )
        except:
            fails.append(mode)
            pass
    print(len(fails))
    #df = df_final.loc[~df_final.index.duplicated(keep='last')]
    
    return df_final
```

```python
df = build_table()
```

Determines the number of wins in each dataset

```python
number_of_wins = np.zeros(11)
for i in range(1402,2830):
    filter2 = df["Method"] != 'bandits_recent'
    filter3 = df["Method"] != 'best_model'
    try:
        filter1 = df["Dataset"] == 'M3F'+str(i)
        df_new = df.where(filter1 & filter2 & filter3)
        ind = df_new[df_new["Dataset"] == 'M3F'+str(i)].MAPE.idxmin()
        number_of_wins[ind] += 1
    except:
        pass
print(number_of_wins)
```
